```python
'''Hunter Bogwill hw13 program'''
import sqlite3
connection = sqlite3.connect('books.db')
import pandas as pd
```

```python
# Select all authorsâ€™ last names from the  table in descending order.
pd.options.display.max_columns = 10 
pd.read_sql('SELECT last FROM authors ORDER BY last DESC', connection)
```

```python
# Select all book titles from the  table in ascending order.
pd.read_sql('SELECT title FROM titles ORDER BY title ASC', connection)
```

```python
# Use an INNER JOIN to select all the books for a specific author. 
# Include the title, copyright year and ISBN. 
# Order the information alphabetically by title.

pd.read_sql("""SELECT first, last, title, copyright, author_ISBN.ISBN FROM titles 
               INNER JOIN author_ISBN ON titles.ISBN = author_ISBN.ISBN
               INNER JOIN authors ON author_ISBN.id = authors.id
               WHERE first LIKE 'Abbey' ORDER BY title""", 
            connection)
```

```python
# Insert a new author Sue Red into the authors table.
cursor = connection.cursor()
cursor = cursor.execute("""INSERT INTO authors (first, last) VALUES ('Sue', 'Red')""")
pd.read_sql('SELECT * FROM authors', connection, index_col=['id'])
```
